#### General Stock Data from Yahoo (not official API)

```python
import pandas_datareader.data as web
```

```python
aapl_df = web.DataReader('AAPL', 'yahoo', start='2020-01-01', end='2021-01-01')
```

```python
aapl_df
```

# FRED

10-Year Breakeven Inflation Rate (T10YIE)
